#### Trabalhando com Series

```python
# Criando uma lista
data = [1, 2, 3, 4, 5]
data
```

```python
# Atribuindo a lista a uma series
s = pd.Series(data)
s
```

```python
# Criar uma lista com nomes para atribuir ao índice da Series
index = ['Linha' + str(i) for i in range(5)]
```

```python
# Mostra o resultado da lista
index
```

```python
# Atribuindo a lista ao índice
s = pd.Series(data = data, index = index)
```

```python
s
```

```python
# Fazendo a mesma coisa usando a criação de um dicionário
data = {'Linha' + str(i) : i + 1 for i in range(5)}
data
```

```python
s = pd.Series(data)
s
```

```python
# Somando um número a cada linha
s1 = s + 2
s1
```

```python
# Somando uma serie a outra (que tenha os mesmos index)
s2 = s + s1
s2

# Caso haja uma linha com index diferente, o resultado será NaN
```
