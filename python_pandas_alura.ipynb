```python
# Criar uma lista
data = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
data
```

```python
# Fazendo o mesmo procedimento utilizando um dicionário
data = {'Coluna0': {'Linha0': 1, 'Linha1': 4, 'Linha2': 7},
        'Coluna1': {'Linha0': 2, 'Linha1': 5, 'Linha2': 8},
        'Coluna2': {'Linha0': 3, 'Linha1': 6, 'Linha2': 9}}
data
```

```python
# Criar o dataframe utilizando o dicionário
df2 = pd.DataFrame(data)
df2
```

```python
# Criar o dicionário por meio de tuplas
# Criar uma lista com tuplas
data = [(1, 2, 3), (4, 5, 6), (7, 8, 9)]
data
```

```python
# Criando o dataframe com a lista de tuplas
df3 = pd.DataFrame(data = data, index = index, columns = columns)
df3
```

```python
# Começar a concatenar dataframes
# Modificar o conteúdo
df1[df1 > 0] = 'A'
df1
```

```python
# Modificar o conteúdo
df2[df2 > 0] = 'B'
df2
```

```python
# Modificar o conteúdo
df3[df3 > 0] = 'C'
df3
```

```python
# Concatenar os dataframes utilizando como parâmetro as colunas
df4 = pd.concat([df1, df2, df3])
df4
```

```python
# Concatenar os dataframes utilizando como parâmetro as linhas
df4 = pd.concat([df1, df2, df3], axis = 1)
df4
```

#### Criando relatório de Análise - Imóveis Residenciais

```python
import pandas as pd
dados = pd.read_csv('dataset/aluguel.csv', sep = ';')
```

```python
dados.head(10)
```

```python
# Verifica todos os tipos de imóveis pela eliminação de dados duplicados na coluna Tipo
list(dados['Tipo'].drop_duplicates())
```

```python
# Criação de uma lista apenas com os tipos de imóveis da categoria Residencial
residencial = [
    'Quitinete', 
    'Casa',
    'Apartamento',
    'Casa de Condomínio',
    'Casa de Vila']
```

```python
# Utilização da função isin() para criação de uma Seies booleana
# que irá verificar se os imóveis no dataset "dados" são da categoria Residencial
dados['Tipo'].isin(residencial)
# Mostrar os 10 primeiros
dados['Tipo'].isin(residencial).head(10)
```

```python
# Criar uma lista de imóveis residenciais na variável selecao
selecao = dados['Tipo'].isin(residencial)
selecao.head()
```

```python
# Criação de um novo dataframe apenas com os registros que possuam a marcação True
dados_residencial = dados[selecao]
dados_residencial
```

```python
# Para verificar se o novo dataframe contém apenas imóveis residenciais
# usaremos a função drop-duplicates()

# Geração de uma lista com os valores únicos na coluna "Tipo"
list(dados_residencial['Tipo'].drop_duplicates())
```

```python
# Verificação da quantidade de linhas do novo dataframe
dados_residencial.shape[0]
```

```python
# Verificação da quantidade de linhas do dataframe original
dados.shape[0]
```

```python
# Com a exclusão dos imóveis não residenciais, o index do novo dataframe não está ordenado
# Refazer o index do novo dataframe para números sequenciais
dados_residencial.index = range(dados_residencial.shape[0])
dados_residencial.head(10)
```
